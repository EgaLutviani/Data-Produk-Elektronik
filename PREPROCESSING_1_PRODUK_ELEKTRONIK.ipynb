```python
#menghapus data yang negatif
data = data[data['harga'] > 0]

#mengganti missing value pada kolom harga
data['harga']=data['harga'].fillna(data['harga'].median())
```

a) Identifikasi outlier dalam dataset

   *Jawab*:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Kolom numerik yang ingin dicek outliernya
num_cols = ['harga', 'rating', 'jumlah_ulasan'] # Removed 'negara_asal' as it's not numerical

# Visualisasi Boxplot
plt.figure(figsize=(15, 8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i) # Adjusted subplot layout for 3 columns
    sns.boxplot(y=data[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)  # Kuartil 1
    Q3 = data[column].quantile(0.75)  # Kuartil 3
    IQR = Q3 - Q1  # Rentang interkuartil
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]
# Cek jumlah outlier untuk setiap kolom numerik
for col in num_cols: # Now only iterates over numerical columns
    outliers = detect_outliers_iqr(data, col)
    print(f"{col}: {len(outliers)} outliers")
```

Dari output diatas tampak bahwa kolom harga, rating, dan jumlah_ulasan terlihat sudah seimbang, sehingga tidak diperlukan penanganan terhadap outliter.

```python
# Buat dictionary untuk mapping negara ke region
country_to_region = {
    'Japan': 'Asia', 'Canada': 'America', 'South Korea': 'Asia', 'Frace': 'Europe', 'USA': 'America', 'Germany': 'Europr', 'China': 'Asia', 'India': 'Asia', 'UK': 'Europe',

}

# Assuming your column is named 'negara_asal', change 'country' to 'negara_asal'
data['country_region'] = data['negara_asal'].map(country_to_region).fillna('Others')

# One-Hot Encoding pada region
data = pd.get_dummies(data, columns=['country_region'], prefix='region')
```

b) Bandingkan berbagai metode encoding dan pilih metode yang paling sesuai untuk setiap kolom. Jelaskan alasan pemilihan metode tersebut.

*Jawab:*

**Label Encoding**: mengubah kategori menjadi angka berdasarkan urutan tertentu dan cocok untuk data dengan hierarki, seperti tingkat pendidikan (SMA < S1 < S2 < S3).

**One-Hot Encoding**: mengonversi kategori menjadi vektor biner, sehingga lebih sesuai untuk data nominal yang tidak memiliki urutan, seperti negara asal dan jenis kelamin.

**Target Encoding**: menggantikan kategori dengan rata-rata nilai dari variabel target dan dapat diterapkan pada kolom seperti kode wilayah jika memiliki korelasi dengan target. Namun, metode ini berisiko menyebabkan data leakage.

**Frequency Encoding**: mengonversi kategori menjadi frekuensi kemunculannya dalam dataset, sehingga cocok untuk data dengan banyak kategori unik, seperti kode pos.

Memilih metode encoding yang tepat sangat penting agar model dapat bekerja secara optimal tanpa kehilangan informasi yang relevan.
